```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
netflix_movies = netflix_df[netflix_df["type"] == 'Movie']
movies_90s = netflix_movies[(netflix_movies["release_year"] >= 1990) & (netflix_movies["release_year"] <2000)]
```

```python
plt.hist(movies_90s['duration'], bins=20)
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.title('Distributions of durations for movies released in the 1990s')
plt.show()
```

```python
movies_action = movies_90s[movies_90s["genre"] == 'Action']

short_movie_count = 0

for label, row in movies_action.iterrows() :
    if row["duration"] < 90 :
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count

print(short_movie_count)
```
